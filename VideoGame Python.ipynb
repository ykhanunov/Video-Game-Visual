```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
grouped = df.groupby('Genre').apply(lambda x: x.nlargest(1, 'Global_Sales'))

fig, ax = plt.subplots(figsize=(10, 5))

fig = sns.barplot(data = grouped, x = 'Name', y = 'Global_Sales', hue = 'Genre').set(xlabel ="Video Game Name", ylabel = "Global Sales by Millions", title ='Top Global Selling Video Game in Each Genre')

plt.xticks(rotation=-85)

```
