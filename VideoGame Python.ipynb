```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
grouped = df.groupby('Genre').apply(lambda x: x.nlargest(2, 'Global_Sales'))

fig = plt.subplots(figsize=(12, 6))
fig = sns.barplot(x = 'Name', y = 'Global_Sales', data = grouped, hue = 'Genre', dodge=False).set(xlabel ="Video Game Name", ylabel = "Global Sales by Millions", title ='Top Two Global Selling Video Game in Each Genre')
plt.xticks(rotation=-85)

plt.show()

```
